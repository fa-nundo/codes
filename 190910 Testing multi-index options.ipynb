# Testing MultiIndex options

For this code, i wanted to be able to move/choose index. I'm using Financial datas of Google and Microsoft.<br /><br />
- I get data as usually, and tried MultiIndex function. But because of this multi-level dataframe, i didn't reach what i wanted to.
- Then i tried using stack() and reindex() but didn't worked
- The solution i found is getting data for each companies (no multi-level dataframe) and concate dataframes. 

A function can probably do the job (and improve the code), with a list of companies we want.<br />
In the part 1.1, the transpose() function offers a satisfying result, but i wanted to check options i had.<br />
Same in part 1.2 for the stack() function.

```python
# Import librairies & Set
import pandas as pd
import numpy as np
import datetime as dt
from pandas_datareader import data

pd.set_option('display.float_format', '{:20,.2f}'.format)
```

## First try (double companies)

```python
company = ['MSFT', 'GOOG']
start = "2010-01-01"
end = "2019-08-13"

stocks = data.DataReader(name = company, data_source = "yahoo", start = start, end = end)
stocks.head()
```

```python
df = stocks.transpose()
df.head()
```

```python
index = pd.MultiIndex.from_product([['GOOG', 'MSFT'],
                                    ['High', 'Low', 'Open', 'Close', 'Volume', 'Adj Close']], 
                                   names=['State', 'Direction'])
df = pd.DataFrame(index=index,data=df.values, columns=df.columns)
df.head()
```

## Third try (reduce dimension, companies one by one)

```python
company = 'GOOG' # without brackets
start = "2010-01-01"
end = "2019-08-13"

stocksss = data.DataReader(name = company, data_source = "yahoo", start = start, end = end)
stocksss.head()
```

```python
ml = stocksss.transpose()
index = pd.MultiIndex.from_product([['GOOG'], 
                                    ['High', 'Low', 'Open', 'Close', 'Volume', 'Adj Close']], 
                                    names=['State', 'Direction'])
df = pd.DataFrame(index=index,data=ml.values, columns=ml.columns)
df.head()
```

```python
company = 'MSFT'
start = "2010-01-01"
end = "2019-08-13"

stocksss = data.DataReader(name = company, data_source = "yahoo", start = start, end = end)
ml = stocksss.transpose()
index = pd.MultiIndex.from_product([['MSFT'], 
                                    ['High', 'Low', 'Open', 'Close', 'Volume', 'Adj Close']], 
                                    names=['State', 'Direction'])
dff = pd.DataFrame(index=index,data=ml.values, columns=ml.columns)
dff.head()
```

## Ok, perfect. Let's concat

```python
frames = [df, dff]
result = pd.concat(frames)
result
```
